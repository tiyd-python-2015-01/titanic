```python
train = training_data.copy()

female = (train.Sex == "female")
male = (train.Sex == "male")
c = (train.Embarked == "C")
q = (train.Embarked == "Q")
s = (train.Embarked == "S")
c1 = (train.Pclass == 1)
c2 = (train.Pclass == 2)
c3 = (train.Pclass == 3)
live = (train.Survived == 1)
die = (train.Survived == 0)
```

```python
"""Calculate chance of surviving based on criteria"""
survival_probability = train.Survived.mean()
total_survivors = train.Survived.sum()
total_passengers = len(train.index.unique())
fcc1l = len(train[female & c & c1 & live].index.unique()) / total_survivors
fcc2l = len(train[female & c & c2 & live].index.unique()) / total_survivors
fcc3l = len(train[female & c & c3 & live].index.unique()) / total_survivors
mcc1l = len(train[male & c & c1 & live].index.unique()) / total_survivors
mcc2l = len(train[male & c & c2 & live].index.unique()) / total_survivors
mcc3l = len(train[male & c & c3 & live].index.unique()) / total_survivors
fqc1l = len(train[female & q & c1 & live].index.unique()) / total_survivors
fqc2l = len(train[female & q & c2 & live].index.unique()) / total_survivors
fqc3l = len(train[female & q & c3 & live].index.unique()) / total_survivors
mqc1l = len(train[male & q & c1 & live].index.unique()) / total_survivors
mqc2l = len(train[male & q & c2 & live].index.unique()) / total_survivors
mqc3l = len(train[male & q & c3 & live].index.unique()) / total_survivors
fsc1l = len(train[female & s & c1 & live].index.unique()) / total_survivors
fsc2l = len(train[female & s & c2 & live].index.unique()) / total_survivors
fsc3l = len(train[female & s & c3 & live].index.unique()) / total_survivors
msc1l = len(train[male & s & c1 & live].index.unique()) / total_survivors
msc2l = len(train[male & s & c2 & live].index.unique()) / total_survivors
msc3l = len(train[male & s & c3 & live].index.unique()) / total_survivors
```

```python
"""Caclulate chance of not surviving based on criteria"""
fcc1d = len(train[female & c & c1 & die].index.unique()) / (total_passengers - total_survivors)
fcc2d = len(train[female & c & c2 & die].index.unique()) / (total_passengers - total_survivors)
fcc3d = len(train[female & c & c3 & die].index.unique()) / (total_passengers - total_survivors)
mcc1d = len(train[male & c & c1 & die].index.unique()) / (total_passengers - total_survivors)
mcc2d = len(train[male & c & c2 & die].index.unique()) / (total_passengers - total_survivors)
mcc3d = len(train[male & c & c3 & die].index.unique()) / (total_passengers - total_survivors)
fqc1d = len(train[female & q & c1 & die].index.unique()) / (total_passengers - total_survivors)
fqc2d = len(train[female & q & c2 & die].index.unique()) / (total_passengers - total_survivors)
fqc3d = len(train[female & q & c3 & die].index.unique()) / (total_passengers - total_survivors)
mqc1d = len(train[male & q & c1 & die].index.unique()) / (total_passengers - total_survivors)
mqc2d = len(train[male & q & c2 & die].index.unique()) / (total_passengers - total_survivors)
mqc3d = len(train[male & q & c3 & die].index.unique()) / (total_passengers - total_survivors)
fsc1d = len(train[female & s & c1 & die].index.unique()) / (total_passengers - total_survivors)
fsc2d = len(train[female & s & c2 & die].index.unique()) / (total_passengers - total_survivors)
fsc3d = len(train[female & s & c3 & die].index.unique()) / (total_passengers - total_survivors)
msc1d = len(train[male & s & c1 & die].index.unique()) / (total_passengers - total_survivors)
msc2d = len(train[male & s & c2 & die].index.unique()) / (total_passengers - total_survivors)
msc3d = len(train[male & s & c3 & die].index.unique()) / (total_passengers - total_survivors)
```

```python
"""Calculate proportion of the population represented by each passenger"""
fcc1prob = len(train[female & c & c1].index.unique()) / total_passengers
fcc2prob = len(train[female & c & c2].index.unique()) / total_passengers
fcc3prob = len(train[female & c & c3].index.unique()) / total_passengers
mcc1prob = len(train[male & c & c1].index.unique()) / total_passengers
mcc2prob = len(train[male & c & c2].index.unique()) / total_passengers
mcc3prob = len(train[male & c & c3].index.unique()) / total_passengers
fqc1prob = len(train[female & q & c1].index.unique()) / total_passengers
fqc2prob = len(train[female & q & c2].index.unique()) / total_passengers
fqc3prob = len(train[female & q & c3].index.unique()) / total_passengers
mqc1prob = len(train[male & q & c1].index.unique()) / total_passengers
mqc2prob = len(train[male & q & c2].index.unique()) / total_passengers
mqc3prob = len(train[male & q & c3].index.unique()) / total_passengers
fsc1prob = len(train[female & s & c1].index.unique()) / total_passengers
fsc2prob = len(train[female & s & c2].index.unique()) / total_passengers
fsc3prob = len(train[female & s & c3].index.unique()) / total_passengers
msc1prob = len(train[male & s & c1].index.unique()) / total_passengers
msc2prob = len(train[male & s & c2].index.unique()) / total_passengers
msc3prob = len(train[male & s & c3].index.unique()) / total_passengers
```

```python
# train.survival_prob *= survival_probability
"""Aggregate survival probability of each passenger"""
train["survival_prob"] = 1
train.loc[female & c & c1, "survival_prob"] *= fcc1l
train.loc[female & c & c2, "survival_prob"] *= fcc2l
train.loc[female & c & c3, "survival_prob"] *= fcc3l
train.loc[male & c & c1, "survival_prob"] *= mcc1l
train.loc[male & c & c2, "survival_prob"] *= mcc2l
train.loc[male & c & c3, "survival_prob"] *= mcc3l

train.loc[female & q & c1, "survival_prob"] *= fqc1l
train.loc[female & q & c2, "survival_prob"] *= fqc2l
train.loc[female & q & c3, "survival_prob"] *= fqc3l
train.loc[male & q & c1, "survival_prob"] *= mqc1l
train.loc[male & q & c2, "survival_prob"] *= mqc2l
train.loc[male & q & c3, "survival_prob"] *= mqc3l

train.loc[female & s & c1, "survival_prob"] *= fsc1l
train.loc[female & s & c2, "survival_prob"] *= fsc2l
train.loc[female & s & c3, "survival_prob"] *= fsc3l
train.loc[male & s & c1, "survival_prob"] *= msc1l
train.loc[male & s & c2,"survival_prob"] *= msc2l
train.loc[male & s & c3, "survival_prob"] *= msc3l
```

```python
"""Aggregate death probability of each passenger"""
train["death_prob"] = 1
train.loc[female & c & c1, "death_prob"] *= fcc1d
train.loc[female & c & c2, "death_prob"] *= fcc2d
train.loc[female & c & c3, "death_prob"] *= fcc3d
train.loc[male & c & c1, "death_prob"] *= mcc1d
train.loc[male & c & c2, "death_prob"] *= mcc2d
train.loc[male & c & c3, "death_prob"] *= mcc3d

train.loc[female & q & c1, "death_prob"] *= fqc1d
train.loc[female & q & c2, "death_prob"] *= fqc2d
train.loc[female & q & c3, "death_prob"] *= fqc3d
train.loc[male & q & c1, "death_prob"] *= mqc1d
train.loc[male & q & c2, "death_prob"] *= mqc2d
train.loc[male & q & c3, "death_prob"] *= mqc3d

train.loc[female & s & c1, "death_prob"] *= fsc1d
train.loc[female & s & c2, "death_prob"] *= fsc2d
train.loc[female & s & c3, "death_prob"] *= fsc3d
train.loc[male & s & c1, "death_prob"] *= msc1d
train.loc[male & s & c2, "death_prob"] *= msc2d
train.loc[male & s & c3, "death_prob"] *= msc3d
```

```python
train["final"] = (train["survival_prob"] * survival_probability) / ((train["survival_prob"] * survival_probability) + 
                                                                    (train["death_prob"] * (1 - survival_probability)))
```
